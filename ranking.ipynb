```python
def create_candidate_text(candidate):

    headline = candidate["profile"].get(
        "headline",
        ""
    )

    summary = candidate["profile"].get(
        "summary",
        ""
    )

    skills = " ".join([
        skill["name"]
        for skill in candidate["skills"]
    ])

    career = " ".join([
        job["description"]
        for job in candidate["career_history"]
    ])

    education = " ".join([
        edu["field_of_study"]
        for edu in candidate["education"]
    ])

    return f"""
    {headline}

    {summary}

    Skills:
    {skills}

    Career:
    {career}

    Education:
    {education}
    """
```

```python
candidate_features = []

for candidate in candidates:

    candidate_features.append({

        "candidate_id":
        candidate["candidate_id"],

        "candidate_text":
        create_candidate_text(
            candidate
        ),

        "experience":
        candidate["profile"][
            "years_of_experience"
        ],

        "open_to_work":
        candidate["redrob_signals"][
            "open_to_work_flag"
        ],

        "github_score":
        candidate["redrob_signals"][
            "github_activity_score"
        ],

        "response_rate":
        candidate["redrob_signals"][
            "recruiter_response_rate"
        ]
    })
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

print("Model Loaded")
```

```python
job_description = """
Senior AI Engineer

Production experience with embeddings
Vector databases
Hybrid search
Retrieval systems
Ranking systems
Python

Evaluation Metrics:
NDCG
MRR
MAP

5-9 years experience

Product company experience

Fine-tuning
LoRA
Learning to Rank

Pune
Noida
"""
```

```python
jd_embedding = model.encode(
    job_description
)
```

```python
# df = df.sample(
#     10000,
#     random_state=42
# )
```

```python
candidate_texts = df[
    "candidate_text"
].tolist()

print(
    "Total Candidate Texts:",
    len(candidate_texts)
)
```

```python
candidate_embeddings = model.encode(
    candidate_texts,
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

scores = cosine_similarity(
    [jd_embedding],
    candidate_embeddings
)[0]
```

```python
df["semantic_score"] = scores

print(
    df[
        [
            "candidate_id",
            "semantic_score"
        ]
    ].head()
)
```

```python
semantic_df = df.sort_values(
    by="semantic_score",
    ascending=False
)

semantic_df.head(10)
```

```python
def calculate_behavior_score(row):

    score = 0

    # Open to work
    if row["open_to_work"]:
        score += 30

    # GitHub Activity
    github = row["github_score"]

    if github > 80:
        score += 30

    elif github > 60:
        score += 20

    elif github > 40:
        score += 10

    # Recruiter Response Rate
    response = row["response_rate"]

    score += response * 40

    return round(score, 2)
```

```python
semantic_df["behavior_score"] = semantic_df.apply(
    calculate_behavior_score,
    axis=1
)
```

```python
semantic_df[
    [
        "candidate_id",
        "github_score",
        "response_rate",
        "behavior_score"
    ]
].head()
```

```python
def experience_score(exp):

    if 6 <= exp <= 8:
        return 100

    elif 5 <= exp <= 9:
        return 80

    elif 4 <= exp <= 10:
        return 50

    return 10
```

```python
semantic_df["experience_score"] = semantic_df[
    "experience"
].apply(experience_score)
```

```python
semantic_df["final_score"] = (

      semantic_df["semantic_score"] * 60

    + semantic_df["behavior_score"] * 0.25

    + semantic_df["experience_score"] * 0.15

)
```

```python
semantic_df = semantic_df.sort_values(
    by=[
        "final_score",
        "candidate_id"
    ],
    ascending=[
        False,
        True
    ]
)
```

```python
semantic_df[
    [
        "candidate_id",
        "semantic_score",
        "behavior_score",
        "experience_score",
        "final_score"
    ]
].head(20)
```

```python
def generate_reasoning(row):

    reasons = []

    if row["semantic_score"] > 0.60:
        reasons.append(
            "Strong match for retrieval, ranking and search systems"
        )

    elif row["semantic_score"] > 0.50:
        reasons.append(
            "Good semantic alignment with AI Engineer role"
        )

    if row["experience_score"] >= 80:
        reasons.append(
            f"{row['experience']} years relevant experience"
        )

    if row["open_to_work"]:
        reasons.append(
            "Open to work"
        )

    if row["github_score"] > 60:
        reasons.append(
            "Strong GitHub activity"
        )

    if row["response_rate"] > 0.50:
        reasons.append(
            "High recruiter response rate"
        )

    return "; ".join(reasons)
```

```python
semantic_df["reasoning"] = semantic_df.apply(
    generate_reasoning,
    axis=1
)
```

```python
top100 = semantic_df.head(100).copy()
```

```python
submission = top100[
    [
        "candidate_id",
        "rank",
        "final_score",
        "reasoning"
    ]
].copy()
```

```python
career_keywords = [
    "retrieval",
    "ranking",
    "search",
    "recommendation",
    "matching",
    "embeddings",
    "vector",
    "relevance",
    "ndcg",
    "mrr",
    "map",
    "evaluation",
    "hybrid search",
    "vector database"
]
```

```python
def get_career_score(candidate):

    text = " ".join([
        job["description"]
        for job in candidate["career_history"]
    ]).lower()

    score = 0

    for keyword in career_keywords:

        if keyword in text:
            score += 10

    return min(score, 100)
```

```python
career_scores = {}

for candidate in candidates:

    career_scores[
        candidate["candidate_id"]
    ] = get_career_score(
        candidate
    )
```

```python
def advanced_behavior_score(candidate):

    s = candidate["redrob_signals"]

    score = 0

    # Open to work
    if s["open_to_work_flag"]:
        score += 20

    # Recruiter response
    score += s["recruiter_response_rate"] * 20

    # Interview completion
    score += s["interview_completion_rate"] * 15

    # GitHub
    github = s["github_activity_score"]

    if github > 0:
        score += min(github / 5, 15)

    # Recruiter interest
    score += min(
        s["saved_by_recruiters_30d"] * 2,
        10
    )

    # Search visibility
    score += min(
        s["search_appearance_30d"] / 20,
        10
    )

    # Short notice period
    notice = s["notice_period_days"]

    if notice <= 30:
        score += 10

    elif notice <= 60:
        score += 5

    # Verified profile
    if s["verified_email"]:
        score += 2

    if s["verified_phone"]:
        score += 2

    return round(score, 2)
```

```python
advanced_scores = {}

for candidate in candidates:

    advanced_scores[
        candidate["candidate_id"]
    ] = advanced_behavior_score(
        candidate
    )
```

```python
semantic_df["advanced_behavior_score"] = (
    semantic_df["candidate_id"]
    .map(advanced_scores)
)
```
